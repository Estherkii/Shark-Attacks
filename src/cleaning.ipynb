# 2. Cleaning "Species"

```python
#cleaning the species column

species_unique = attacks['species'].unique()
species_unique
```

```python
attacks['species'] = attacks['species'].str.lower()
```

```python
attacks['species'] = attacks['species'].str.replace(r'[^a-zA-Z0-9 ]', '', regex=True)
```

```python
def species(input_str):
    
    input_str = input_str.lower() 

    if re.search(r'\bwhite\b', input_str):
        return 'white'
    if re.search(r'\btiger\b', input_str):
        return 'tiger'
    if re.search(r'\blemon\b', input_str):
        return 'lemon'
    if re.search(r'\bbull\b', input_str):
        return 'bull'
    if re.search(r'\b(?:grey|reef)\b', input_str):
        return 'grey reef'
    if re.search(r'\bwobbegong\b', input_str):
        return 'wobbegong'
    if re.search(r'\bshark\b', input_str):
        return 'blacktip'
    if re.search(r'\bgalapagos\b', input_str):
        return 'galapagos'
    if re.search(r'\bnurse\b', input_str):
        return 'nurse'
    if re.search(r'\bspinner\b', input_str):
        return 'spinner'
    if re.search(r'\bporbeagle\b', input_str):
        return 'porbeagle'
    if re.search(r'\bwhitetip\b', input_str):
        return 'whitetip'
    if re.search(r'\bblue\b', input_str):
        return 'blue'
    if re.search(r'\b(?:broadnose|seven|gill)\b', input_str):
        return 'broadnose'
    if re.search(r'\bshovelnose\b', input_str):
        return 'shovelnose'
    if re.search(r'\bsilvertip\b', input_str):
        return 'silvertip'
    if re.search(r'\bronze\b', input_str):
        return 'bronze'
    if re.search(r'\sandbar\b', input_str):
        return 'sandbar'
    
    return 'other'

attacks['species'] = attacks['species'].apply(species)
```

#  3. Cleaning "Country"

```python
#cleaning the value 'country'

country_clean = attacks['country'].unique()

country_clean
```

```python
def country(input_str):
    input_str = input_str.lower() 
    
    for country in pycountry.countries:
        if re.search(r'\b{}\b'.format(country.name.lower()), input_str):
            return country.name.upper()

    return 'other'

attacks['country'] = attacks['country'].apply(country)
```

#  3. Cleaning "Activity"

```python
attacks['activity'] = attacks['activity'].str.lower()
```

```python
attacks['activity'] = attacks['activity'].str.replace(r'[^a-zA-Z0-9 ]', '', regex=True)
```

```python
def activity(input_str):
    if re.search(r'\b(swimming|bathing)\b', input_str):
        return 'swimming'
    if re.search(r'\bfishing\b', input_str, re.IGNORECASE):
        return 'fishing'
    if re.search(r'\bsurfing\b', input_str, re.IGNORECASE):
        return 'surfing'
    if re.search(r'\bboat|yatch|ship\b', input_str, re.IGNORECASE):
        return 'sailing'
    if re.search(r'\bshark\b', input_str, re.IGNORECASE):
        return 'shark interacting'
    if re.search(r'\bdiving|dived?\b', input_str, re.IGNORECASE):
        return 'diving'
    if re.search(r'\b(kayaking|canoe|paddling)\b', input_str):
        return 'kayaking'
    return 'other'

attacks['activity'] = attacks['activity'].apply(activity)
```

#  Cleaning "Time"

```python
def clean_and_standardize_time(time_str):
    # Remove all characters that are not numbers, colons, or 'h'
    cleaned_time = re.sub(r'[^0-9:h]', '', str(time_str))

    # Remove any trailing 'h'
    cleaned_time = cleaned_time.rstrip('h')

    # Handle words or sentences, convert them to NaN
    if re.search(r'\b\w+\b', cleaned_time):
        return np.nan

    # Split the time using 'h' and ':'
    time_parts = re.split(r'[h:]+', cleaned_time)

    if len(time_parts) >= 2:
        # Ensure that there are at least two time parts (HH and MM)
        hh = time_parts[0].zfill(2)
        mm = time_parts[1].zfill(2)
        return f"{hh}:{mm}"
    else:
        return np.nan
```
